```python
# - load matplotlib and numpy
import matplotlib.pyplot as plt
import numpy as np
```

```python
# - load data
filename = '../data/CO2_emissions_Americas_1960-2017_nopound.txt'
data = np.loadtxt(filename, dtype=str)

# - save country names as a separate array !! NOTE: you don't want the first column as that is for the year...
countries = data[0,1:]

# - save years and CO2 emissions as separate numerical arrays (1d arrays)
years = np.asarray(data[1:,0], dtype=float)
CO2 = np.asarray(data[1:,1:], dtype=float)

# - print 1st 5 lines of countries, years and CO2 (for CO2, just the 1st country)
print(countries[:5])
print(years[:5])
print(CO2[:5,0])
```

```python
# - print all countries
print(countries)
```

```python
# - get the index for the US
us_index = np.argwhere(countries=='USA')
print(us_index)

# - now get the index for Belize
belize_index = np.argwhere(countries=='Belize')
print(belize_index)

# - now get the index for Chile
chile_index = np.argwhere(countries=='Chile')
print(chile_index)
```

```python
# - print out the 1st 5 rows (i.e. 1st 5 years) of the data for each country
print(CO2[:5,us_index[0]])
print(CO2[:5,belize_index[0]])
print(CO2[:5,chile_index[0]])
```

```python
# - plot the US
plt.plot(years,CO2[:,us_index[0]], label='USA')

# - plot Belize
plt.plot(years,CO2[:,belize_index[0]], label='Belize')

# - plot Chile
plt.plot(years,CO2[:,chile_index[0]], label='Chile')

# - plot legend
plt.legend()

# - add labels and title
plt.xlabel('Years')
plt.ylabel('CO2 Emissions')
plt.title('CO2 Emissions per country (1960-2017)')

# - show plot
plt.show()
```

EXTRA if you are familiar with Python

```python
# - plot a bar plot for the CO2 emissions in 2017 for the first 10 countries in the Americas CO2 emissions file

# - set-up x-axis array for barplot
ncountries = 10
x_scalars = np.arange(ncountries)
print(x_scalars)

# - set-up y axis array for barplot
CO2_2017 = CO2[-1,:ncountries]
print(CO2_2017)

# - plot bar
plt.bar(x_scalars, CO2_2017)

# - plot xtick labels
plt.xticks(x_scalars, countries[:ncountries], rotation=45)

# - plot axes labels and title
plt.xlabel('Country')
plt.ylabel('CO2 emissions')
plt.title('2017 CO2 emissions per country')

# - plot show()
plt.show()
```
