```python
# - load pandas, matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# - load data
filename = '../data/fluxnet/FLUXNET_US-Srm_Daily_1989-2014.csv'
data_df = pd.read_csv(filename)
```

#### Question: which plot would be useful for showing Day vs Nighttime temperature? 

**Plot and show that plot now, with a legend, labels and a title.**

```python
# - plot the 1st year's day and nighttime temperature and given them a label
plt.plot(data_df.TA_DAY[:365], label='Day')
plt.plot(data_df.TA_NIGHT[:365], label='Night')

# - legend
plt.legend()

# - labels and titles
plt.xlabel('DOY')
plt.ylabel('Air Temperature ($^{o}$C)')
plt.title('Day vs nighttime temperature for US-Wkg Fluxnet site')

# - show
plt.show()
plt.close()
```

#### Question: What if we wanted to know if there is a relationship between daily air temperature and VPD - which plot would be useful for showing that relationship? 
**Plot and show that plot now, with labels and a title.**

```python
# - plot the scatter plot between TA and VPD
plt.scatter(data_df.TA, data_df.VPD)

# - labels and title
plt.xlabel('Daily air temp. ($^{o}$C})')
plt.ylabel('VPD')
plt.title('Relationship between Tair and VPD')

# - show and close
plt.show()
plt.close()
```
