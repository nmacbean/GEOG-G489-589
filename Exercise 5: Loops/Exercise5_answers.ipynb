```python
# - import libraries
import numpy as np
import pandas as pd
import glob
```

```python
# - create filelist
#filelist = glob.glob('C:\\Users\\nmacbean\\Downloads\\data\\*.csv')
filelist = glob.glob('../data/fluxnet/data/*.csv')
print(filelist)
```

```python
# - create array of zeros for precipiation sums
allsite_precip_sum = np.zeros(len(filelist))
```

```python
# - loop over files and read in data
for nf, f in enumerate(filelist):
    
    # - read each file into a pandas dataframe
    df = pd.read_csv(f)
    
    # - print simple statistics
    print(df.describe())
    
    # - save precipiation sums
    allsite_precip_sum[nf] = np.sum(df.P)
```

```python
# - print out the precipiation sums for all sites
print(allsite_precip_sum)
```
