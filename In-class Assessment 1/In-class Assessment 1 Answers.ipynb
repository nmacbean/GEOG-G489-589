```python
# - import libraries
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
# - get a list of all the files
filelist = glob.glob('../data/fluxnet/data/*.csv')
print(filelist)
```

```python
# - now loop over all the files
for f in filelist:
    
    # - read data into a pandas df and print out the filename
    df = pd.read_csv(f)
    print(f)
    
    # - plot a scatter plot between P and VPD and add labels and a title
    plt.scatter(df.P, df.VPD)
    plt.xlabel('Precipitation')
    plt.ylabel('Vapor Pressure Deficit')
    plt.title('P-VPD relationship for '+f)
    plt.show()
    plt.close()
```

Scientific question answers:
1. What is the relationship between vapor pressure deficit (VPD) and precipitation (P) at each site? (I.e. what happens to VPD when P increases?)

ANSWER: As P increases, VPD decreases following a logarithmically decreasing function.

2. Does this relationship between VPD and P differ between the California sites and the Arizona sites?

ANSWER: No, although there is a greater spread of VPD values at non-zero but low precipitation values of ~5-20 mm/day.
